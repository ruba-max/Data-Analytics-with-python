```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"E:\Data Visualization\Books_Data_Clean.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
df = df[df["Publishing Year"] > 1900]
```

```python
df.isna().sum()
```

```python
df.dropna(subset = "Book Name",inplace = True)
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

```python
df.nunique()
```

```python
plt.hist(df["Publishing Year"])
plt.xlabel("Publishing Year")
plt.ylabel("Frequency")
plt.title("Distribution of Publishing Year")
plt.show()
```

```python
df["genre"].value_counts().plot(kind = "bar")
df.xlabel("Genre")
df.ylabel("Number of books")
df.title("Number of Books in Each Genre")
df.show()
```

```python
df["genre"].value_counts().plot(kind = "bar")
df.xlabel("Genre")
df.ylabel("Number of books")
df.title("Number of Books in Each Genre")
df.show()
```

```python
df.groupby("Author")["Book_average_rating"].mean().sort_values(ascending = False)
```

```python
sns.boxplot(x = "genre", y = "Book_ratings_count", data = df)
plt.xlabel("Genre")
plt.ylabel("Book_ratings_count")
plt.title("Box plot of Book Rating Count of Each Book")
plt.show()
```

```python
total_gross_by_author = df.groupby("Author")["gross sales"].sum()
```

```python
total_gross_by_author.sort_values(ascending = False).head(10).plot(kind = "bar")
plt.xlabel("Author")
plt.ylabel("gross sales")
plt.title("Total Gross Sales for each author")
plt.show()
```

```python
sns.boxplot(x="Author_Rating",y="units sold", data = df)
plt.xlabel("Author_Rating")
plt.ylabel("units sold")
plt.title("Box plot of Units sold of each author")
plt.show()
```

```python
df.groupby("Publishing Year")["units sold"].sum().plot(kind = "line", marker= "o")
plt.xlabel("Publishing Year")
plt.ylabel("Total units sold")
plt.title("Total Units sold over the years")
plt.show()
```
