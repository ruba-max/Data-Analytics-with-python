```python
sns.countplot(data = data, x = "Sex")
plt.title("Distribution by Gender")
plt.show()
```

```python
sns.histplot(data = data, x = "Age", bins = 20, kde = True)
plt.title("Age Distribution")
plt.show()
```

```python
sns.histplot(data = data, x = "Height", bins = 20, kde = True)
plt.title("Height Distribution")
plt.show()
```

```python
sns.histplot(data = data, x = "Weight", bins = 20, kde = True)
plt.title("Distribution of weight")
plt.show()
```

```python
sns.countplot(data = data, x = "Medal")
plt.title("Medals Distribution")
plt.show()
```

```python
sns.countplot(data = data, x = "Year", hue = "Medal")
plt.title("Medal distribution over Years")
plt.xticks(rotation = 75)
plt.show()
```

```python
average_age_by_years = data.groupby("Year")["Age"].mean()
print(average_age_by_years)
```

```python
sport_median_height = data.groupby("Sport")["Height"].median()
print(sport_median_height.max())
```

```python
sport_median_height[sport_median_height == 190.0] 
```

```python
country_gender_count = data.groupby(["NOC","Sex"])["ID"].count()
print(country_gender_count)
```

```python
# Which country has the most gold medals
country_gold_medals = data[data["Medal"]== "Gold"].groupby("NOC")["Medal"].count()
print(country_gold_medals.max())
```

```python
country_gold_medals[country_gold_medals == 747]
```

```python
sport_gender_avg_weight = data.groupby(["Sport", "Sex"])["Weight"].mean()
print(sport_gender_avg_weight)
```

```python
print(sport_gender_avg_weight["Wrestling"]["F"])
```

```python
sport_event_count = data.groupby("Sport")["Event"].nunique().sort_values(ascending = False)
sport_event_count.plot(kind = "bar")
plt.title("Number of Unique Events per sport ")
plt.xlabel("Sport")
plt.ylabel("Number of Unique Events")
plt.xticks(rotation = 75)
```

```python
year_participant_count = data.groupby("Year")["ID"].nunique()
year_participant_count.plot(kind = "bar")
plt.title("Number of participant over the years")
plt.xlabel("Year")
plt.ylabel("Number of participant")
plt.show()
```

```python
# Visualize the average of the participant in different countries
data.head(1)
```

```python
avg_age_by_country= data.groupby("NOC")["Age"].mean().sort_values(ascending = False).head(10)
avg_age_by_country.plot(kind ="bar", color = "pink")
plt.title("Top 10 country's average age of participant")
plt.xlabel("Country")
plt.ylabel("Age")
plt.xticks(rotation = 45)
plt.show()
```

```python
sns.boxplot(data = data, x = "Season", y = "Age",hue = "Season", palette={"Summer":"blue", "Winter": "yellow"})
plt.title("Distribution of ages by season")
plt.show()
```

```python
sns.violinplot(data = data, x= "Medal",y ="Height", hue = "Medal",palette = "Set2")
plt.title("Distribution of Heights by Medal")
plt.xlabel("Medal")
plt.ylabel("Height")
plt.show()
```

```python
# Finding out the most medal winning country
most_medals_wining_country = data["NOC"].value_counts().idxmax()
print("Most medal winning country: ", most_medals_wining_country)
```

```python
# Finding out the athelete with highest weight
highest_weight_athelete = data[data["Weight"] == data["Weight"].max()]
print("Highest weight athelete: ")
print(highest_weight_athelete[["Id", "Name", "Sport"]])
```

```python
tallest_athelete = data[data["Height"] == data["Height"].max()]
print("Tallest athelete: ")
print(tallest_athelete[["ID", "Name", "Sport"]])
```

```python
# height and weight combination of athelete with medal winning informations
sns.scatterplot(data = data, x = "Height", y = "Weight", hue = "Medal", palette = "Set2")
plt.title("Athelete Height and Weight by Medal Status")
plt.xlabel("Height")
plt.ylabel("Weight")
plt.show()
```

```python
medals_by_country_year = data.pivot_table(index = "NOC", columns = "Year", values = "Medal", aggfunc = "count")
sns.heatmap(medals_by_country_year, cmap = "YlGnBu", linewidths = 0.5)
plt.title("Medal count by country and year")
plt.xlabel("Year")
plt.ylabel("Country")
plt.xticks(rotation = 45)
plt.show()
```
